```python
from matplotlib import pyplot as plt

import numpy as np
np.random.seed(30)
import random
random.seed(30)
```

```python
from sklearn.datasets import fetch_20newsgroups

categories = ['comp.sys.ibm.pc.hardware', 'comp.graphics',
'comp.sys.mac.hardware', 'comp.os.ms-windows.misc',
'rec.autos', 'rec.motorcycles',
'rec.sport.baseball', 'rec.sport.hockey']
dataset = fetch_20newsgroups(subset='all', categories=categories,shuffle=True, random_state=42)
```

# Question 1

```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
t1 = CountVectorizer(min_df=3,stop_words='english')
X_counts = t1.fit_transform(dataset.data)

t2 = TfidfTransformer()
X_tfidf = t2.fit_transform(X_counts)

print(" X_count ", X_counts.shape)
print(" X_tfidf ", X_tfidf.shape)
```

```python
dataset.bintarget = np.zeros_like(dataset.target)
for y in [0, 1, 2, 3]:
    dataset.bintarget[np.argwhere(dataset.target == y)] = 1
for y in [4, 5, 6, 7]:
    dataset.bintarget[np.argwhere(dataset.target == y)] = 0
class_names = ['Recreational Activity','Computer Technology']
print(dataset.bintarget.shape)
dataset.target=dataset.bintarget
```

```python
dataset.bintarget.sum()
```

# Question 2

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, init='k-means++',n_init=30, max_iter=1000, random_state=0)
#when we change the max_iter to 500 or 1500,the resulat does not change
kmeans.fit(X_tfidf)
X_predict = kmeans.predict(X_tfidf)
X_predict.shape
print(" X_predict ", X_predict.shape)
```

```python
from sklearn.metrics.cluster import contingency_matrix

con_matrix=contingency_matrix(dataset.target,X_predict)
print(con_matrix)
```

```python
import itertools
def plot_contingency_matrix(cm, classes,
                          normalize=False,
                          title='Contingency matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the contingency matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum()
        print("Normalized Contingency matrix")
    else:
        print('Unnormalization Contingency Matrix')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
plt.figure()
plot_contingency_matrix(con_matrix, classes=class_names, normalize=True,
                      title='Normalized Contingency Matrix')

plt.show()

plt.figure()
plot_contingency_matrix(con_matrix, classes=class_names, normalize=False,
                      title='Unnormalization Contingency Matrix')

plt.show()
```

```python
r_list=[1,2,3,5,10,20,50,100,300]
```

```python
def svd_kmeans(r):
    svd = TruncatedSVD(n_components=r, random_state=42)
    X_reduced = svd.fit_transform(X_tfidf)
    km = KMeans(n_clusters=2, init='k-means++',n_init=30, max_iter=1000, random_state=0)
    km.fit(X_reduced)
    return [metrics.homogeneity_score(dataset.target, km.labels_),metrics.completeness_score(dataset.target, km.labels_),metrics.v_measure_score(dataset.target, km.labels_),metrics.adjusted_rand_score(dataset.target, km.labels_),metrics.adjusted_mutual_info_score(dataset.target, km.labels_)]
    
```

```python
from sklearn.decomposition import NMF
def NMF_kmeans(r):
    model = NMF(n_components=r, init='random', random_state=42)
    X_reduced = model.fit_transform(X_tfidf)
    km = KMeans(n_clusters=2, init='k-means++',n_init=30, max_iter=1000, random_state=0)
    km.fit(X_reduced)
    return [metrics.homogeneity_score(dataset.target, km.labels_),metrics.completeness_score(dataset.target, km.labels_),metrics.v_measure_score(dataset.target, km.labels_),metrics.adjusted_rand_score(dataset.target, km.labels_),metrics.adjusted_mutual_info_score(dataset.target, km.labels_)]
```

```python
def plot_measure(r_list,reduced_type):
    y=[]
    if reduced_type==1:
        print("PCA")
        for i in r_list:
            y.append(svd_kmeans(i))
        y = np.asarray(y).T.tolist()
        plt.figure()
        plt.plot(y[0])
        plt.xlabel("r components")
        plt.ylabel("Homogeneity")
        plt.xticks(range(10),r_list)
        plt.figure()
        plt.plot(y[1])
        plt.xlabel("r components")
        plt.ylabel("Completeness")
        plt.xticks(range(10),r_list)
        plt.figure()
        plt.plot(y[2])
        plt.xlabel("r components")
        plt.ylabel("V-measure")
        plt.xticks(range(10),r_list)
        plt.figure()
        plt.plot(y[3])
        plt.xlabel("r components")
        plt.ylabel("Adjusted Rand-Index")
        plt.xticks(range(10),r_list)
        plt.figure()
        plt.plot(y[4])
        plt.xlabel("r components")
        plt.ylabel("Adjusted Mutual Info Score")
        plt.xticks(range(10),r_list)        
    else:
        print("NMF")
        for i in r_list:
            y.append(NMF_kmeans(i))
        y = np.asarray(y).T.tolist()
        plt.figure()
        plt.plot(y[0])
        plt.xlabel("r components")
        plt.ylabel("Homogeneity")
        plt.xticks(range(10),r_list)
        plt.figure()
        plt.plot(y[1])
        plt.xlabel("r components")
        plt.ylabel("Completeness")
        plt.xticks(range(10),r_list)
        plt.figure()
        plt.plot(y[2])
        plt.xlabel("r components")
        plt.ylabel("V-measure")
        plt.xticks(range(10),r_list)
        plt.figure()
        plt.plot(y[3])
        plt.xlabel("r components")
        plt.ylabel("Adjusted Rand-Index")
        plt.xticks(range(10),r_list)
        plt.figure()
        plt.plot(y[4])
        plt.xlabel("r components")
        plt.ylabel("Adjusted Mutual Info Score")
        plt.xticks(range(10),r_list)
```

```python
def reduce_dim(svd_or_NMF, r, X):
    if svd_or_NMF==1:
        svd = TruncatedSVD(n_components=r, random_state=42)
        X_reduced = svd.fit_transform(X)
        title='(SVD)'
    else:
        model = NMF(n_components=r, init='random', random_state=42)
        X_reduced = model.fit_transform(X)
        title='(NMF)'
    if r>2:
        svd_vis=TruncatedSVD(n_components=2, random_state=42)
        X_reduced=svd_vid.fit_transform(X_reduced)
    return X_reduced, title
def visualization(X_rd, km, title):
    plt.figure()
    color1=[]

    plt.scatter(X_rd[:,0], X_rd[:,1], s=1, c=0.05*km.labels_)
    title1='clustring label'+title
    plt.xlabel(title1)
    plt.figure()
    color2=[]

    plt.scatter(X_rd[:,0], X_rd[:,1], s=1,c=0.05*dataset.target)
    title1='Ground truth class label'+title
    plt.xlabel(title1)
    plt.show()
    print("Homogeneity: %0.3f" % metrics.homogeneity_score(dataset.target, km.labels_))
    print("Completeness: %0.3f" % metrics.completeness_score(dataset.target, km.labels_))
    print("V-measure: %0.3f" % metrics.v_measure_score(dataset.target, km.labels_))
    print("Adjusted Rand-Index: %.3f"
      % metrics.adjusted_rand_score(dataset.target, km.labels_))
    print("adjusted_mutual_info_score: %0.3f"
      % metrics.adjusted_mutual_info_score(dataset.target, km.labels_))
```

# SVD

```python
r_svd=2
X_rd_svd, title1=reduce_dim(1, r_svd, X_tfidf)
km = KMeans(n_clusters=2, init='k-means++',n_init=30, max_iter=1000, random_state=0)
km.fit(X_rd_svd)
visualization(X_rd_svd, km, title1)
```

# NMF

```python
r_nmf=2
X_rd_nmf, title2=reduce_dim(0, r_nmf, X_tfidf)
km = KMeans(n_clusters=2, init='k-means++',n_init=30, max_iter=1000, random_state=0)
km.fit(X_rd_nmf)
visualization(X_rd_nmf, km, title2)
```

# Question 8, 9, 10

```python
def Scaling_features(X):
    X/=np.std(X,axis=0,ddof=1)
    return X
def log_transformation(X):
    return np.log(abs(X)+0.01)
```

# SVD scaling features

```python
r_svd=2
X_rd_svd, title1=reduce_dim(1, r_svd, X_tfidf)
X = Scaling_features(X_rd_svd)
km = KMeans(n_clusters=2, init='k-means++',n_init=30, max_iter=1000, random_state=0)
km.fit(X) 
visualization(X,km,'(SVD) Scaling Features')
```

# NMF scaling features

```python
r_nmf=2
X_rd_nmf, title2=reduce_dim(0, r_nmf, X_tfidf)
X = Scaling_features(X_rd_nmf)
km = KMeans(n_clusters=2, init='k-means++',n_init=30, max_iter=1000, random_state=0)
km.fit(X) 
visualization(X,km,'(nmf) normalize variance')
```

# NMF Log Transformation

```python
r_nmf=2
X_rd_nmf, title2=reduce_dim(0, r_nmf, X_tfidf)
X = log_transformation(X_rd_nmf)
km = KMeans(n_clusters=2, init='k-means++',n_init=30, max_iter=1000, random_state=0)
km.fit(X) 
visualization(X,km,'(NMF) log transformation')
```

# NMF Scaling featuring and then Log Transformation

```python
r_nmf=2
X_rd_nmf, title2=reduce_dim(0, r_nmf, X_tfidf)
X = log_transformation(Scaling_features(X_rd_nmf))
km = KMeans(n_clusters=2, init='k-means++',n_init=30, max_iter=1000, random_state=0)
km.fit(X) 
visualization(X,km,'(NMF) Scaling featuring and then transform')
```
