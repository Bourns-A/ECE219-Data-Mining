```python
def Heatmap(df):
    corr = df.corr()
    plt.figure(figsize = (8, 8))
    sns.heatmap(corr,
               xticklabels = corr.columns,
               yticklabels = corr.columns)
def Histogram(df):
    df.hist(figsize = (15, 15), layout = (6, 3), ec = 'black', grid = False, column = ['duration', 'height',
    'width', 'bitrate', 'framerate', 'i', 'p', 'b', 'frames', 'i_size', 'p_size', 'b_size', 'size', 'o_bitrate', 
    'o_framerate', 'o_width', 'o_height'])
    
def Boxplot(df):
    for col in ['codec', 'o_codec']:
        df.boxplot(by = col, column = 'utime', figsize = (6, 6))
```
